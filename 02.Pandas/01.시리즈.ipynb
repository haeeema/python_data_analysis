- 시리즈와 딕셔너리

```python
'서울' in s, '대전' in s
```

```python
for key, value in s.items():
    print(f'{key}: {value}')
```

```python
d = {"서울": 9631482, "부산": 3393191, "인천": 2632035, "대전": 1490158}
s2 = pd.Series(d)
s2
```

```python
# 인덱스의 순서를 정하고 싶으면, 인덱스를 별도로 설정해야 함
s2 = pd.Series(d, index=['서울', '부산', '인천', '대전'])
s2
```

- 인덱스 기반 연산

```python
ds = s - s2
ds
```

```python
ds.notna(), ds.notnull()
```

```python
ds[ds.notna()]
```

```python
# 증가율(% 단위)
rs = (s - s2) / s2 * 100
rs = rs[rs.notna()]
rs
```

- 데이터의 갱신, 추가, 삭제

```python
rs['부산'] = 1.64
rs[1] = 2.83
rs.인천 = 9.82
rs
```

```python
# rs[3] = '광주' // err
```

```python
rs['광주'] = 1.51   # ❗추가하는 경우 rs[3], rs.광주 는 사용할 수 없음
rs
```

```python
del rs['부산']      # ❗삭제하는 경우 rs[0], rs.부산은 사용할 수 없음
```

```python
rs
```

- Titanic data로 확인

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head()      # 앞의 다섯줄 보여주기
```

```python
survived = titanic['survived']
age = titanic['age']
```

```python
survived.isna().sum()   # 결과가 0이므로 결측치가 없음
```

```python
age.isna().sum()        # 177개의 결측치가 있음
```

- 결측치 처리 방법
    - 제거
    - 대표값(중앙값, 평균, 최빈값)으로 대체
    - 임의의 값으로 대체

```python
# 제거
ds.dropna(inplace=True)
ds
```

```python
# 평균으로 대체
age = age.fillna(age.mean()).astype(int)
age.head()
```
